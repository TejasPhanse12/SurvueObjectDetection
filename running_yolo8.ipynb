```python
from ultralytics import YOLO
import os

# Load OpenVINO model
model = YOLO('runs/detect/yolo8nnew3/weights/best_int8_openvino_model')

# Run validation
results = model.val(data='data.yaml', imgsz=1280)

# Size
openvino_path = 'runs/detect/yolo8nnew3/weights/best_int8_openvino_model'
total_size = sum(os.path.getsize(os.path.join(openvino_path, f)) for f in os.listdir(openvino_path))
print(f"Model Size:   {total_size / (1024*1024):.2f} MB")
```
